```python
tickers = ['T','BAC','ARES','AAPL','AMZN']
```

```python
short = []
e = []

for i in range(len(tickers)):
    url = 'https://research.tdameritrade.com/grid/public/research/stocks/summary?symbol={}'.format(tickers[i])
    r = requests.get(url)
    soup = bs4.BeautifulSoup(r.text,'html')
    info = soup.find('div',{'id':'layout-body'}).find('div',{'class':'col-xs-6 table-data-wrapper'}).find_all('div',{'class':'col-xs-6'})
    short_interest = info[-1].find_all('li')[-1].find('dd').text
    short.append(short_interest)
    
    url = 'https://finance.yahoo.com/quote/{a}/profile?p={a}'.format(a=tickers[i])
    r = requests.get(url)
    soup = bs4.BeautifulSoup(r.text,'html')
    employees = soup.find('div',{'class':'asset-profile-container'}).find('div',{'class':'Mb(25px)'}).find_all('p')[-1].find_all('span',{'class':'Fw(600)'})[-1].text
    e.append(employees)
```

```python
pd.DataFrame({'Tickers':tickers,'Short Interest(% of float)':short,'Employees':e})
```

```python
industries = ['technology','consumer-goods','finance','industrial-conglomerates','utilities','industrials','materials','health-and-science']
```

```python
for i in range(len(industries)):
    url = 'https://www.cnbc.com/{}/'.format(industries[i])
    r = requests.get(url)
    soup = bs4.BeautifulSoup(r.text,'html')
    headline = soup.find('div',{'class':'PageBuilder-col-9 PageBuilder-col'}).find('div',{'class':'Card-textContent'})
    title = headline.text
    link = headline.find('a')['href']
    print('{}: {}\n Link: {}'.format(industries[i].capitalize(), title, link))
```
